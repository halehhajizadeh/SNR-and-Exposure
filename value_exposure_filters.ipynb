```python
import numpy as np
import matplotlib.pylab as plt
from mpl_toolkits.axes_grid.inset_locator import (inset_axes, InsetPosition, mark_inset)
```

# Signal vs exposur time

```python
exposure_time = [1,5,10,15,20,25,30,45,60,90,120,150,180,210,240,270,300,360,420,480,540,600,900]

value_gprime = [1000,1051,1123,1120,1115,1242,1228,1476,1476,1978,2362,2104,2432,2686,3379,3526,3727,4238,5323,4777,6319,6441,8010]
value_iprime = [1000,1062,1118,1179,1180,1202,1292,1374,1432,1604,1764,1872,1961,2136,2268,2402,2682,2915,3218,3691,4385,5140,7401]
value_rprime = [1061,1119,1231,1363,1255,1266,1375,1508,1757,2120,2652,2706,2956,3409,3608,3950,4022,4987,5943,6441,8339,8918,10656]
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
ax1.plot(exposure_time, value_gprime, 'go', label='g_prime')
ax1.plot(exposure_time, value_gprime, 'g--')
ax1.plot(exposure_time, value_iprime, 'b^', label='i_prime')
ax1.plot(exposure_time, value_iprime, 'b-.')
ax1.plot(exposure_time, value_rprime, 'rx', label='r_prime')
ax1.plot(exposure_time, value_rprime, 'r:')
ax1.set_title('Value vs Exposure time with 3 Filters', fontsize=20)
ax1.set_xlabel('Exposure time', fontsize=15)
ax1.set_ylabel('Value',fontsize=15)
ax1.legend(loc=4)

# Create a set of inset Axes: these should fill the bounding box allocated to
# them.
ax2 = plt.axes([0,0,0,0])
# Manually set the position and relative size of the inset axes within ax1
ip = InsetPosition(ax1, [0.05,0.5,0.3,0.45])
ax2.set_axes_locator(ip)

ax2.plot(exposure_time, value_gprime, 'go', label='SNR_Star1')
ax2.plot(exposure_time, value_gprime, 'g--')
ax2.plot(exposure_time, value_iprime, 'b^', label='SNR_Star2')
ax2.plot(exposure_time, value_iprime, 'b-.')
ax2.plot(exposure_time, value_rprime, 'rx', label='SNR_Star3')
ax2.plot(exposure_time, value_rprime, 'r:')
ax2.set_xlim(0,80)
ax2.set_ylim(900,3000)
```
